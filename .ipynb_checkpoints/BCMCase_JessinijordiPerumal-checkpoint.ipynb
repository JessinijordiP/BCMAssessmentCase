```python
import pandas as pd
import numpy as np
```

```python
hr_2015 = pd.read_csv(r'Data Sets\2015_Happiness_report.csv')
hr_2016 = pd.read_csv(r'Data Sets\2016_Happiness_report.csv')
hr_2017 = pd.read_csv(r'Data Sets\2017_Happiness_report.csv')
hr_2018 = pd.read_csv(r'Data Sets\2018_Happiness_report.csv')
hr_2019 = pd.read_csv(r'Data Sets\2019_Happiness_report.csv')
country_codes = pd.read_json(r'Data Sets\countries_continents_codes_flags_url.json')
```

```python
hr_2015.head()
```

```python
hr_2017.head()
```

```python
hr_2015.info()
```

```python
hr_2015.drop('Standard Error', axis = 1, inplace = True)
```

```python
hr_2015.columns = ['Country', 'Happiness score', 'GDP per capita', 'Family', 'Healthy life expectancy', 'Freedom to make life choices', 'Perceptions of corruption', 'Generosity', 'Social support']
hr_2015.head()
```

```python
hr_2016.drop(['Lower Confidence Interval','Upper Confidence Interval'], axis = 1, inplace = True)
hr_2016
```

```python
hr_2016.columns = ['Country', 'Happiness score', 'GDP per capita', 'Family', 'Healthy life expectancy', 'Freedom to make life choices', 'Perceptions of corruption', 'Generosity', 'Social support']
hr_2016.head()
```

```python
hr_2017.drop(['Whisker.high','Whisker.low'], axis = 1, inplace = True)

hr_2017
```

```python
hr_2017.columns = ['Country', 'Happiness score', 'GDP per capita', 'Family', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity' ,'Perceptions of corruption', 'Social support']
hr_2017
```

```python
hr_2018.rename({'Country or region' :'Country' ,'Score': 'Happiness score'}, inplace= True, axis = 1 )
hr_2018
```

```python
hr_2019.rename({'Country or region' :'Country' ,'Score': 'Happiness score'}, inplace= True, axis = 1 )
hr_2019
```

```python
# Adding Year Columns --------- 
hr_2015['Year'] = 2015
hr_2016['Year'] = 2016
hr_2017['Year'] = 2017
hr_2018['Year'] = 2017
hr_2019['Year'] = 2019
```

```python
hr_2015.shape, hr_2016.shape, hr_2017.shape, hr_2018.shape, hr_2019.shape
```

```python
country_codes.head()
```

```python
country_codes.columns = ['Country', 'Image file', 'Country Url', 'Alpha 2', 'Alpha 3', 'Country code', 'Iso_3161_2', 'Region', 'Sub region', 'Intermediate region', 'Region code', 'Sub region code','Intermediate region code']
```

```python
#droping extra columns from Country codes
country_codes = country_codes[['Country', 'Country Url', 'Region', 'Region code']]
```

```python
#Region to upper case
country_codes['Region'] = country_codes['Region'].str.upper()
```

```python
country_codes
```

```python
# merging the report data with country details ------- 
hr_2015_final = hr_2015.merge(country_codes, on = 'Country', how = 'left' )
hr_2016_final = hr_2016.merge(country_codes, on = 'Country', how = 'left' )
hr_2017_final = hr_2017.merge(country_codes, on = 'Country', how = 'left' )
hr_2018_final = hr_2018.merge(country_codes, on = 'Country', how = 'left' )
hr_2019_final = hr_2019.merge(country_codes, on = 'Country', how = 'left' )
```

```python
hr_2015_final
```

```python
data_list = [hr_2015_final, hr_2016_final, hr_2017_final, hr_2018_final, hr_2019_final]
```

```python
# Happiness Status for all data frames
for i in data_list :
    i['Happiness status'] = (
    np.select(
        condlist=[i['Happiness score'] > 5.6, (i['Happiness score'] < 5.6)&(i['Happiness score'] >= 2.6), i['Happiness score'] < 2.6], 
        choicelist=['Green', 'Amber', 'Red']))
```

```python
# Creating the overall rank per year
for i in data_list :
    i['Overall Rank'] = i['Happiness score'].rank(ascending = False)
```

```python
# Creating the ranking per region
for i in data_list:
    i["Rank per region"] = i.groupby("Region")["Happiness score"].rank("dense", ascending=False)
```

```python
hr_2015_final.columns, len(hr_2015_final.columns)
# same as per task 3,  only - country code is an extra variable.
```

```python
data = pd.concat(data_list,ignore_index= True, sort= False)
```

```python
# Final Data
data
```

```python
# Extracting csv file of data
data.to_csv('FinalData.csv')
```

```python
# Task 4 - Empty dataframe
task_4 = pd.DataFrame()
```

```python
# Creating Highest Rank column
task_4['Highest Rank'] = data.groupby('Country')['Overall Rank'].max()
```

```python
# Creating Lowest Rank column
task_4['Lowest Rank'] = data.groupby('Country')['Overall Rank'].min()
```

```python
# Creating Lowest Rank column
task_4['Highest Happiness Score'] = data.groupby('Country')['Happiness score'].max()
```

```python
# Creating Lowest Rank column
task_4['Lowest Happiness Score'] = data.groupby('Country')['Happiness score'].min()
```

```python
# Resetting Index for dataframe
task_4 = task_4.reset_index()
```

```python
# Task 4 completed
task_4
```

```python
# Extracting json file for task 4
task_4.to_json('finaldata.json')
```
